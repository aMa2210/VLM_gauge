```python
import os
import pandas as pd


folder_list = [
    r'Results/1.- Analog Dial Gaugegemini-3-pro-preview',
    r'Results/2.- Analog Depth Gaugegemini-3-pro-preview',
]

results = {}
all_values = []

for folder in folder_list:

    for file_name in os.listdir(folder):
        if file_name.endswith('.xlsx'):
            file_path = os.path.join(folder, file_name)

            try:
                df = pd.read_excel(file_path)

                if 'total_duration_sec' in df.columns:
                    values = df['total_duration_sec'].dropna()
                    if len(values) == 1:
                        all_values.append(values.iloc[0])
                    elif len(values) == 0:
                        print(f"there is no total_duration_sec in {file_path}")
                    else:
                        raise ValueError(f"{file_path} has more than one total_duration_sec")
                    
                else:
                    print('error code2322')

            except Exception as e:
                print(f"failed loading: {file_path}, error: {e}")

if all_values:
    avg_value = sum(all_values) / len(all_values)
else:
    avg_value = None

print(avg_value)

```
